```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets as widgets
%matplotlib inline
from IPython.display import display
%matplotlib notebook

Models = ['GRU','SRNN (16 1)', 'SRNN (8 2)', 'SRNN (4 3)', 'SRNN (2 8)', 'DCCNN']
Dataset = ['Yelp 2013', 'Yelp 2014', 'Yelp 2015', 'Yelp_P', 'Amazon_F', 'Amazon_P']

def myplot_dataset(dataset,model):
    data = pd.read_csv('srnn_data.csv')
    data.index = data.Dataset
    data = data.drop(['Dataset'],axis = 1)
    #l = list(dataset)
    #print(l)
    mydata = data[data.Model == model].loc[list(dataset)]
    mydata.drop(columns='Model', inplace=True)
    mydata = mydata.transpose()
    #print(mydata)
    mydata.plot.bar()
        
    
def myplot_model(model, dataset):
    data = pd.read_csv('srnn_data.csv')
    data.index = data.Model
    data = data.drop(['Model'],axis = 1)
    #dataset_index = Dataset[dataset.index]
    mydata = data[data.Dataset == dataset].loc[list(model)]
    mydata.drop(columns='Dataset', inplace=True)
    mydata = mydata.transpose()
    mydata.plot.bar()
    #plt.figure(figsize=(10,6))
    
def choice(select):
    if select== 'Dataset':
        dataset = widgets.SelectMultiple(options=Dataset, value=['Yelp 2013'], rows=6, description='Dataset')
        model = widgets.RadioButtons(options=Models, description='Models')
        i=widgets.HBox([ model, dataset])
        out = widgets.interactive_output(myplot_dataset, {'dataset':dataset ,'model':model})
        display(i,out)
    else:
        dataset = widgets.RadioButtons(options=Dataset, description='Dataset')
        model = widgets.SelectMultiple(options=Models, value=['GRU'], rows=6, description='Models')
        i=widgets.HBox([dataset, model])
        out = widgets.interactive_output(myplot_model, {'model':model, 'dataset':dataset})
        display(i,out)

select = widgets.RadioButtons(options=['Dataset', 'Model'], description='Select:')
widgets.interactive(choice, select = select)
```
